```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns = df.columns.str.normalize('NFKC').str.strip().str.replace(r'\s+', '_', regex=True)

```

```python
df.rename(columns={'orderid':'order_id','orderdate':'order_date','shipmode':'ship_mode','postalcode':'postal_code','subcategory':'sub_category','productid':'product_id','costprice':'cost_price','listprice':'list_price','discountpercent':'discount_percent'})
```
